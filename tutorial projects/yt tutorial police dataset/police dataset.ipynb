```python
import pandas as pd
```

```python
df = pd.read_csv('3. Police Data.csv')
```

```python
df
```

## 1) Remove the column that only contains missing value

```python
df.dropna(axis=1, how='all', inplace=True)
df
```

## 2) From speeding, were men or women stopped more often?

```python
df[df['violation'] == 'Speeding']['driver_gender'].value_counts()
```

## 3) Does gender affect who gets searched?

```python
df.groupby('driver_gender')['search_conducted'].sum()
```

## 4) What is the mean stop duration?

```python
df.stop_duration = df.stop_duration.map({'0-15 Min' : 7.5, '16-30 Min' : 23, '30+ Min' : 45})
```

```python
df.head()
```

```python
df.stop_duration.mean()
```
