## Line graph: Gas prices

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
plt.figure(figsize=(8,5))
plt.title('Gas prices over time (in USD)', fontdict={'fontweight':'bold', 'fontsize':18})

#plt.plot(gas.Year, gas.USA, 'b.-', label = 'United States')
#plt.plot(gas.Year, gas.Canada, 'r.-', label = 'Canada')
#plt.plot(gas.Year, gas.Australia, 'g.-', label = 'Australia')

## Select all countrys at once
#for country in gas:
#    if country != 'Year': 
#        plt.plot(gas.Year, gas[country], marker='.', label = country)

## Select specific values at once
countries_to_look = ['Australia', 'USA', 'Canada', 'South Korea']
for country in countries_to_look:
   if country != 'Year': 
       plt.plot(gas.Year, gas[country], marker='.', label = country)

## Start at the beggining, end at the end, but skip every 3 year
plt.xticks(gas.Year[::3]) 

plt.xlabel('Year')
plt.ylabel('US dolars')
plt.legend()
plt.show
```

## Fifa Data

### Histogram

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head(5)
```

```python
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall, bins=bins)
plt.xticks(bins)

plt.title('Distribution of player skills')
plt.ylabel('Number os players')
plt.xlabel('Skill level')
plt.show()
```

### Pie chart

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
fifa['Preferred Foot'].value_counts()

plt.pie(fifa['Preferred Foot'].value_counts(), labels=['Left', 'Right'][::-1], autopct='%.2f %%')
```

```python
fifa.Weight
```

```python
fifa.Weight = [float(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]
```

```python
fifa.Weight = fifa.Weight.apply(lambda x:x*0.45)
```

```python
fifa.Weight
```

```python
light = fifa.loc[fifa.Weight < 56.25].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 56.25) & (fifa.Weight < 67.5)].count()[0]
medium = fifa.loc[(fifa.Weight >= 67.5) & (fifa.Weight < 78.75)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 78.75) & (fifa.Weight < 90)].count()[0]
heavy = fifa.loc[fifa.Weight >= 90].count()[0]

plt.title('Weight distribution of fifa players (kg)')
plt.style.use('ggplot')
weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['Under 56.25', '56.25-67.5', '67.5-78.75', '78.75-90', 'Over 90']
plt.pie(weights, labels = labels, autopct='%.2f %%', explode=(.6,.1,.0,.0,.6))

plt.show()
```

```python
plt.style.use('default')
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']

plt.boxplot([barcelona, madrid], labels=['FC Barcelona','Real Madrid'])
```
